Diagramas

```python
matriz_correlacion = datos.corr().round(2)
plt.figure(figsize=(10, 10))
sns.heatmap(data=matriz_correlacion, annot=True)
```

```python
correlacion=datos.corr()
plt.figure(figsize=(10, 10))

sns.heatmap(correlacion, vmax=.8, linewidths=0.01,
            square=True,annot=True,cmap='YlGnBu',linecolor="white")
plt.title('Correlation between features');
```

```python
plt.figure(figsize=(10, 10))
ax = sns.boxplot(x="Outcome", y="Age", 
                data=datos)

ax = sns.stripplot(x="Outcome", y="Age",
                   data=datos, jitter=True,
                   edgecolor="gray")
```
