## Visualización de datos

### Correlación de los datos:

Se realiza inicialmente una análisis de correlación entre los datos para determinar a cuales se les harán los diagramas correspondientes

```python
corr=diabetes.corr()
plt.figure(figsize=(10, 10))

sns.heatmap(corr, vmax=.8, linewidths=0.01,
            square=True,annot=True,cmap='YlGnBu',linecolor="white")
plt.title('Correlación entre características');
```

### Histograma de la edad:

```python
# set the size of the figure
sns.set(rc={'figure.figsize':(11.7,8.27)})

# plot a histogram showing the distribution of the target values
sns.displot(diabetes['Age'], bins=10)
plt.show()
```

Con el histograma para la edad podemos apreciar que el rango para la edad en esta base de datos ronda mayormente entre los 20 hasta los 40 años.

### Personas con o sin Diabetes: 

```python
fig = plt.figure(figsize=(18,6))      ## To get a figure with proper structure
diabetes.Outcome.value_counts().plot(kind="bar",alpha=0.5)  
```

Se aprecia como la mayoría de las personas en la base de datos no tiene diabetes

### Relación entre los niveles de glucosa y la diabetes:

```python
bx = sns.boxplot(x="Outcome", y="Glucose", 
                data=diabetes)

bx = sns.stripplot(x="Outcome", y="Glucose",
                   data=diabetes, jitter=True,
                   edgecolor="gray")
```

Se puede apreciar en gracias a las cajas de bigotes que las personas con diabetes tienden a tener niveles de glucosa superiores a las personas que no poseen diabetes, confirmando la correlación de Outcome y Glucose

### Relación entre el BMI y la Diabetes

```python
bx = sns.boxplot(x="Outcome", y="BMI", 
                data=diabetes)

bx = sns.stripplot(x="Outcome", y="BMI",
                   data=diabetes, jitter=True,
                   edgecolor="gray")
```

Este valor nos parece indicar que entre mayor sea el indice de mása corporal, mayorr es la probabilidad de padecer diabetes
