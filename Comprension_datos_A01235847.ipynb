Exploraci√≥n de Datos

```python
diabetes = pd.read_csv('diabetes.csv')
```

```python
##Pregnancies

promedio_Pregnancies= diabetes['Pregnancies'].mean()
mediana_Pregnancies =diabetes['Pregnancies'].median()
moda_Pregnancies = diabetes['Pregnancies'].mode()
print("promedio_Pregnancies:",promedio_Pregnancies)
print("Mediana_Pregnancies:",mediana_Pregnancies)
print("Moda_Pregnancies:",moda_Pregnancies)
```

```python
##Glucose

promedio_Glucose= diabetes['Glucose'].mean()
mediana_Glucose =diabetes['Glucose'].median()
moda_Glucose = diabetes['Glucose'].mode()
print("promedio_Glucose:",promedio_Glucose)
print("Mediana_Glucose:",mediana_Glucose)
print("Moda_Glucose:",moda_Glucose)
```

```python
##BloodPressure

promedio_BloodPressure= diabetes['BloodPressure'].mean()
mediana_BloodPressure =diabetes['BloodPressure'].median()
moda_BloodPressure= diabetes['BloodPressure'].mode()
print("promedio_BloodPressure:",promedio_BloodPressure)
print("Mediana_BloodPressure:",mediana_BloodPressure)
print("Moda_BloodPressure:",moda_BloodPressure)
```

```python
##SkinThickness
promedio_SkinThickness= diabetes['SkinThickness'].mean()
mediana_SkinThickness = diabetes['SkinThickness'].median()
moda_SkinThickness = diabetes['SkinThickness'].mode()
print("promedio_SkinThickness:",promedio_SkinThickness)
print("Mediana_SkinThickness:",mediana_SkinThickness)
print("Moda_SkinThickness:",moda_SkinThickness)
```

```python
##Insulin
promedio_Insulin = diabetes['Insulin'].mean()
mediana_Insulin = diabetes['Insulin'].median()
moda_Insulin = diabetes['Insulin'].mode()
print("promedio_Insulin:",promedio_Insulin)
print("Mediana_Insulin:",mediana_Insulin)
print("Moda_Insulin:",moda_Insulin)
```

```python
##BMI
promedio_BMI = diabetes['BMI'].mean()
mediana_BMI = diabetes['BMI'].median()
moda_BMI = diabetes['BMI'].mode()
print("promedio_BMI:",promedio_BMI)
print("Mediana_BMI:",mediana_BMI)
print("Moda_BMI:",moda_BMI)
```

```python
##DiabetesPedigreeFunction

promedio_DiabetesPedigreeFunction = diabetes['DiabetesPedigreeFunction'].mean()
mediana_DiabetesPedigreeFunction = diabetes['DiabetesPedigreeFunction'].median()
moda_DiabetesPedigreeFunction = diabetes['DiabetesPedigreeFunction'].mode()
print("promedio_DiabetesPedigreeFunction:",promedio_DiabetesPedigreeFunction)
print("Mediana_DiabetesPedigreeFunction:",mediana_DiabetesPedigreeFunction)
print("Moda_DiabetesPedigreeFunction:",moda_DiabetesPedigreeFunction)
```

```python
##Age

mean_Age = diabetes['Age'].mean()
median_Age = diabetes['Age'].median()
mode_Age = diabetes['Age'].mode()
print("Mean_Age:",mean_Age)
print("Median_Age:",median_Age)
print("Mode_Age:",mode_Age)
```

```python
##Outcome

mean_Outcome = diabetes['Outcome'].mean()
median_Outcome = diabetes['Outcome'].median()
mode_Outcome = diabetes['Outcome'].mode()
print("Mean_Outcome:",mean_Outcome)
print("Median_Outcome:",median_Outcome)
print("Mode_Outcome:",mode_Outcome)
```

```python

import matplotlib.pyplot as plt 
correlation_matrix = diabetes.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
corr=diabetes.corr()
plt.figure(figsize=(10, 10))

sns.heatmap(corr, vmax=.8, linewidths=0.01,
            square=True,annot=True,cmap='YlGnBu',linecolor="white")
plt.title('Correlation between features')
```
