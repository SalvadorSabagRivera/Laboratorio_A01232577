Eliminacion de variables 

```python
diabetesF = diabetes.drop(['SkinThickness'], axis = 1)
diabetesF = diabetes.drop(['BloodPressure'], axis = 1)

diabetesF.head(10)
```

```python
# Seleccionar dos variables y ver cómo se agrupan
datos1 = diabetesF.loc[:, ['BMI', 'Glucose']]
# KMEANS 1
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(datos1)
labels = kmeans.predict(datos1)

# visualization
plt.scatter(diabetesF['BMI'], diabetesF['Glucose'], c = labels)
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.show()
```

```python
iner_list = np.empty(10)
for i in range(1,10):
    kmeans2 = KMeans(n_clusters=i)
    kmeans2.fit(datos1)
    iner_list[i] = kmeans2.inertia_


plt.plot(range(0,10), iner_list,'-')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
 #Seleccionar dos variables y ver cómo se agrupan
datos1 = diabetesF.loc[:, ['Insulin', 'Age']]
# KMEANS 2
from sklearn.cluster import KMeans
kmeans2 = KMeans(n_clusters=3)
kmeans2.fit(datos1)
labels2 = kmeans2.predict(datos1)
list_clusters = np.unique(labels2)
print("Número de clusters:",len(list_clusters))
print("Cluster Centers:",kmeans2.cluster_centers_)


# visualization
plt.scatter(diabetesF['Insulin'], diabetesF['Age'], c = labels2)
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], c='red', marker='x')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Fare VS Age')
plt.show()
```
