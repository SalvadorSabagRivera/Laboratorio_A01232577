# Preparación de datos

Basándonos en la actividad pasada, podemos determinar que las variables de BloodPressure y SkinThickness no son relevantes para las personas que tienen o no diabetes, por lo que serán removidas

```python
diabetes_prep = diabetes.drop(['BloodPressure','SkinThickness'], axis=1)
```

```python
diabetes_prep.head()
```

```python
#Quitar la característica de la clase
diabetes_prep = diabetes_prep.drop(['Outcome'], axis=1)
```

```python
diabetes_prep.head()
```

```python
# Seleccionar dos variables y ver cómo se agrupan
datos1 = diabetes_prep.loc[:, ['Insulin', 'Glucose']]
from sklearn.cluster import KMeans
# KMEANS 1
kmeans = KMeans(n_clusters=2)
kmeans.fit(datos1)
labels = kmeans.predict(datos1)
# visualization
plt.scatter(diabetes_prep['Insulin'], diabetes_prep['Glucose'], c = labels)
plt.xlabel('Insulin')
plt.ylabel('Glucose')
plt.show()
```

```python
df = pd.DataFrame({'labels':labels, 'Outcome': diabetes['Outcome']})
crosstab = pd.crosstab(df['labels'],df['Outcome'])
crosstab
```

La tabla muestra:  
En el cluster con etiqueta 0 tenemos 421 personas que no tienen diabetes y 182 personas que sí tienen diabetes
En el cluster con etiqueta 1 tenemos 79 personas que no tienes diabetes y 86 personas que sí tienen diabetes

Buscar un valor de k óptimo en base a la inercia, siendo la inercia la suma de las distancias al cuadrado de 
cada objeto del Cluster a su centroide.  
El punto en el que se observa ese cambio brusco en la inercia nos dirá el número óptimo de Clusters a seleccionar para ese data set; 
o dicho de otra manera: el punto que representaría al codo del brazo será el número óptimo de Clusters para ese data set.


```python
iner_list = np.empty(10)
for i in range(1,10):
    kmeans2 = KMeans(n_clusters=i)
    kmeans2.fit(datos1)
    iner_list[i] = kmeans2.inertia_

# iner_list = iner_list%100
# show the best number in graph
plt.plot(range(0,10), iner_list,'-')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
# Seleccionar dos variables y ver cómo se agrupan
datos1 = diabetes_prep.loc[:, ['Insulin', 'Glucose']]
# KMEANS 2
from sklearn.cluster import KMeans
kmeans2 = KMeans(n_clusters=2)
kmeans2.fit(datos1)
labels2 = kmeans2.predict(datos1)
list_clusters = np.unique(labels2)
print("Número de clusters:",len(list_clusters))
print("Cluster Centers:",kmeans2.cluster_centers_)


# visualization
plt.scatter(diabetes_prep['Insulin'], diabetes_prep['Glucose'], c = labels2)
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], c='red', marker='x')
plt.xlabel('Insulin')
plt.ylabel('Glucose')
plt.title('Insulin VS Glucose')
plt.show()
```
