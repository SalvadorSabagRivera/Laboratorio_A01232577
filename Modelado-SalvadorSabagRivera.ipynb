# PreparaciÃ³n de datos

```python
datos_prep = datos.drop(['BloodPressure','SkinThickness','Outcome'], axis=1)
```

```python
datos_prep.head()
```

```python
plt.figure(figsize=(10, 10))
datos1 = datos_prep.loc[:, ['Insulin', 'Glucose']]

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(datos1)
labels = kmeans.predict(datos1)

plt.scatter(datos_prep['Insulin'], datos_prep['Glucose'], c = labels)
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
iner_list = np.empty(10)
for i in range(1,10):
    kmeans2 = KMeans(n_clusters=i)
    kmeans2.fit(datos1)
    iner_list[i] = kmeans2.inertia_

plt.plot(range(0,10), iner_list,'-')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
plt.figure(figsize=(10, 10))
datos1 = datos_prep.loc[:, ['Insulin', 'Glucose']]

from sklearn.cluster import KMeans
kmeans2 = KMeans(n_clusters=3)
kmeans2.fit(datos1)
labels2 = kmeans2.predict(datos1)
list_clusters = np.unique(labels2)
print("NÃºmero de clusters:",len(list_clusters))
print("Cluster Centers:",kmeans2.cluster_centers_)

plt.scatter(datos_prep['Insulin'], datos_prep['Glucose'], c = labels2)
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], c='red', marker='x')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.title('Insulin VS Glucose')
plt.show()
```
